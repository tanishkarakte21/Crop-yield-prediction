```python
def undummy(df):
    category_cols = [col for col in df.columns if col.startswith('Soil_type:_')]
    df['Soil_Type'] = df[category_cols].idxmax(axis=1).str.replace('Soil_type:_', '').astype(int)
    df = df.drop(columns=category_cols)
    return df
```

```python
df
```

```python

new_df_bajra=df[df["Crop:_Bajra"]>=1]
```

```python
new_df_bajra
```

```python
new_df_bajra.columns
```

```python
x_fe=[ 'Temperature',
       'Precipitaion', 'Humidity', 'Soil_type:_chalky', 'Soil_type:_clay',
       'Soil_type:_loamy', 'Soil_type:_peaty', 'Soil_type:_sandy',
       'Soil_type:_silt', 'Soil_type:_silty', 'District:_AHMEDNAGAR',
       'District:_AKOLA', 'District:_AMRAVATI', 'District:_AURANGABAD',
       'District:_BEED', 'District:_BHANDARA', 'District:_BULDHANA',
       'District:_CHANDRAPUR', 'District:_DHULE', 'District:_GADCHIROLI',
       'District:_GONDIA', 'District:_HINGOLI', 'District:_JALGAON',
       'District:_JALNA', 'District:_KOLHAPUR', 'District:_LATUR',
       'District:_NAGPUR', 'District:_NANDED', 'District:_NANDURBAR',
       'District:_NASHIK', 'District:_OSMANABAD', 'District:_PARBHANI',
       'District:_PUNE', 'District:_SANGLI', 'District:_SATARA',
       'District:_SOLAPUR', 'District:_THANE', 'District:_WARDHA',
       'District:_WASHIM', 'District:_YAVATMAL', 'Crop:_Bajra']
```

```python

x=new_df_bajra[x_fe]
```

```python
y_fe=['Yield']
y=new_df_bajra[y_fe]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
x_train
```

```python
x_test
```

## linear regression

```python
from sklearn.linear_model import LinearRegression
model_baj=LinearRegression()
```

```python
model_baj.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
```

```python
model_baj.score(x_train,y_train)
```

```python
y_predict=model_baj.predict(x_test)
```

```python
print(mae(y_test,y_predict))
print(mse(y_test,y_predict))
print(r2_score(y_test,y_predict))
```

```python
model_baj.coef_
```

```python
model_baj.intercept_
```

## random forest

```python
y_pred=model_random.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
```

## decisionTreeRegresson

```python
from sklearn.tree import DecisionTreeRegressor
mode_dec=DecisionTreeRegressor()
mode_dec.fit(x_train,y_train)
y_pred=mode_dec.predict(x_test)
```

## data visualization

```python
import pandas as pd

data=new_df_bajra
columns_to_extract = ['Temperature', 'Precipitaion', 'Humidity',]
new_data = data[columns_to_extract]
```

```python
import matplotlib.pyplot as plt


values = {}
for col in columns_to_extract:
    values[col] = new_data[col].sum()


fig, ax = plt.subplots()
ax.pie(values.values(), labels=values.keys(), autopct='%1.1f%%', startangle=90)
ax.set_title('Data Distribution')

plt.show()
```
